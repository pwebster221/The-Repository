```python
#converting to neo4j
from neo4j import GraphDatabase

class AstroGraph:
    def __init__(self, uri, user, password):
        self.driver = GraphDatabase.driver(uri, auth=(user, password))
        
    def close(self):
        self.driver.close()
    
    def create_user(self, user_id):
        with self.driver.session() as session:
            session.run("MERGE (u:User {id: $user_id})", user_id=user_id)
    
    def create_house(self, house_id, start_degree, end_degree, sign):
        with self.driver.session() as session:
            session.run(
                """
                MERGE (h:House {id: $house_id})
                SET h.start_degree = $start_degree, h.end_degree = $end_degree, h.sign = $sign
                """, house_id=house_id, start_degree=start_degree, end_degree=end_degree, sign=sign
            )
    
    def create_planet(self, name, sign, full_degree, is_retro):
        with self.driver.session() as session:
            session.run(
                """
                MERGE (p:Planet {name: $name})
                SET p.sign = $sign, p.full_degree = $full_degree, p.is_retro = $is_retro
                """, name=name, sign=sign, full_degree=full_degree, is_retro=is_retro
            )
    
    def create_relationships(self, user_id, house_id, planet_name, relationship_type):
        with self.driver.session() as session:
            session.run(
                """
                MATCH (u:User {id: $user_id})
                MATCH (h:House {id: $house_id})
                MATCH (p:Planet {name: $planet_name})
                MERGE (p)-[r:IN_HOUSE {type: $relationship_type}]->(h)
                MERGE (u)-[:HAS_NATAL_PLANET]->(p)
                """, user_id=user_id, house_id=house_id, planet_name=planet_name, relationship_type=relationship_type
            )
    
    def create_aspects(self, aspecting_planet, aspected_planet, aspect_type, orb, diff):
        with self.driver.session() as session:
            session.run(
                """
                MATCH (p1:Planet {name: $aspecting_planet})
                MATCH (p2:Planet {name: $aspected_planet})
                MERGE (p1)-[a:ASPECT {type: $aspect_type, orb: $orb, diff: $diff}]->(p2)
                """, aspecting_planet=aspecting_planet, aspected_planet=aspected_planet, aspect_type=aspect_type, orb=orb, diff=diff
            )

def load_data_into_neo4j(json_data):
    astro_graph = AstroGraph(uri="bolt://localhost:7687", user="neo4j", password="password")
    
    try:
        user_id = "632629"  # Assuming the user ID is consistent
        astro_graph.create_user(user_id)
        
        for house in json_data['houses']:
            house_id = house['house_id']
            start_degree = house['start_degree']
            end_degree = house['end_degree']
            sign = house['sign']
            astro_graph.create_house(house_id, start_degree, end_degree, sign)
            
            for planet in house['planets']:
                name = planet['name']
                planet_sign = planet['sign']
                full_degree = planet['full_degree']
                is_retro = planet['is_retro'] == 'true'
                astro_graph.create_planet(name, planet_sign, full_degree, is_retro)
                astro_graph.create_relationships(user_id, house_id, name, "Natal")
        
        for aspect in json_data['aspects']:
            aspecting_planet = aspect['aspecting_planet']
            aspected_planet = aspect['aspected_planet']
            aspect_type = aspect['type']
            orb = aspect['orb']
            diff = aspect['diff']
            astro_graph.create_aspects(aspecting_planet, aspected_planet, aspect_type, orb, diff)
    
    finally:
        astro_graph.close()

# Example usage
json_data = {
    # Your JSON data goes here
}

load_data_into_neo4j(json_data)
```
